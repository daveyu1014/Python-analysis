```python
import pandas as pd
import csv
```

```python
df = pd.read_csv('netflix_titles.csv')
df.head()
```

### 1. what is the most popular release year for movies on Netflix?

```python
df['type'].unique()
```

```python
df_movie = df[df['type'] == 'Movie']
df_movie.head()
```

```python
#df.groupby(['col1', 'col2']).size().reset_index(name='counts')
df_movie.groupby('release_year').size().reset_index(name='counts').sort_values(by='counts',ascending=False)
```

```python
df_movie['count'] = 1
df_movie.groupby('release_year').sum().reset_index()[['release_year','count']].sort_values('count',ascending=False).head()
```

### 2. What year did netflix add the most content to its platform?

```python
df.info()
```

```python
df['year'] = pd.to_datetime(df['date_added'])
df['year'] = df['year'].dt.year
df.head()
```

```python
#df.groupby(['col1', 'col2']).size().reset_index(name='counts')
df.groupby('year').size().reset_index(name='counts').sort_values(by='counts',ascending=False)
```

### 3. Independent of year, what is the most popular month to add new content?

```python
#df.groupby('year').size().reset_index(name='counts').sort_values(by='counts',ascending=False)
df['added_month'] = pd.to_datetime(df['date_added']).dt.month
df.groupby('added_month').size().reset_index(name='counts').sort_values(by='counts',ascending=False)
```

### 5.Which actor/ actress showup most frequently in the netflix dataset?

```python
#pd.options.display.max_colwidth = 150
df_title_and_cast = df[['title','cast']]
df_title_and_cast = df_title_and_cast.dropna()
```

```python
cast_name_count = {}
for n,cast in enumerate(df_title_and_cast['cast']):
    cast_split =cast.split(', ')
    for name in cast_split:
        cast_name_count[name] = cast_name_count.get(name, 0) +1
        
cast_name_count
```

```python
#sorted(myDict, key= myDict.get, reverse= True)
#dict(sorted(x.items(), key = lambda item: item[i]))
#sorted(cast_name_count, key=cast_name_count.get, reverse=True)
dict(sorted(cast_name_count.items(), key=lambda item:item[1],reverse = True))
```
